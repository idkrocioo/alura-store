### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



# 2. Ventas por categor√≠a

```python
# Agrupo por la columna "Categor√≠a del Producto" y cuento cu√°ntos productos hay por categor√≠a

print("üõí Tienda 1 - Categor√≠as m√°s vendidas:")
print(tienda["Categor√≠a del Producto"].value_counts())
print("\n")

print("üõí Tienda 2 - Categor√≠as m√°s vendidas:")
print(tienda2["Categor√≠a del Producto"].value_counts())
print("\n")

print("üõí Tienda 3 - Categor√≠as m√°s vendidas:")
print(tienda3["Categor√≠a del Producto"].value_counts())
print("\n")

print("üõí Tienda 4 - Categor√≠as m√°s vendidas:")
print(tienda4["Categor√≠a del Producto"].value_counts())
```

# 3. Calificaci√≥n promedio de la tienda


```python
# Calculo el promedio de calificaciones por tienda

promedio_t1 = tienda["Calificaci√≥n"].mean()
promedio_t2 = tienda2["Calificaci√≥n"].mean()
promedio_t3 = tienda3["Calificaci√≥n"].mean()
promedio_t4 = tienda4["Calificaci√≥n"].mean()

print("Calificaci√≥n promedio por tienda:")
print("Tienda 1:", round(promedio_t1, 2))
print("Tienda 2:", round(promedio_t2, 2))
print("Tienda 3:", round(promedio_t3, 2))
print("Tienda 4:", round(promedio_t4, 2))
```

# 4. Productos m√°s y menos vendidos

```python
# Productos m√°s vendidos por tienda
print("Top 5 productos m√°s vendidos - Tienda 1")
print(tienda["Producto"].value_counts().head(5))
print("\n")

print("Top 5 productos m√°s vendidos - Tienda 2")
print(tienda2["Producto"].value_counts().head(5))
print("\n")

print("Top 5 productos m√°s vendidos - Tienda 3")
print(tienda3["Producto"].value_counts().head(5))
print("\n")

print("Top 5 productos m√°s vendidos - Tienda 4")
print(tienda4["Producto"].value_counts().head(5))
```

```python
# Productos menos vendidos por tienda
print("Productos menos vendidos - Tienda 1")
print(tienda["Producto"].value_counts().tail(5))
print("\n")

print("Productos menos vendidos - Tienda 2")
print(tienda2["Producto"].value_counts().tail(5))
print("\n")

print("Productos menos vendidos - Tienda 3")
print(tienda3["Producto"].value_counts().tail(5))
print("\n")

print("Productos menos vendidos - Tienda 4")
print(tienda4["Producto"].value_counts().tail(5))
```

# 5. Env√≠o promedio por tienda

```python
# Calculo el costo promedio de env√≠o por tienda

envio_t1 = tienda["Costo de env√≠o"].mean()
envio_t2 = tienda2["Costo de env√≠o"].mean()
envio_t3 = tienda3["Costo de env√≠o"].mean()
envio_t4 = tienda4["Costo de env√≠o"].mean()

print("Costo promedio de env√≠o por tienda:")
print("Tienda 1:", round(envio_t1, 2))
print("Tienda 2:", round(envio_t2, 2))
print("Tienda 3:", round(envio_t3, 2))
print("Tienda 4:", round(envio_t4, 2))
```
